```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Paths to data
df_movies = pd.read_csv("csv/movies.csv")
df_ratings = pd.read_csv("csv/ratings.csv")

df_movies.head()
```

# Data Analysis (EDA)

The provided data offers valuable insights into user activity and preferences within the movie rating database. With over 58,000 movies and almost 28 million user ratings, the database provides a comprehensive view of which movies are popular and how users react to them.

By analyzing the number of ratings per year, it can be observed that user activity has varied over time. For example, time-series data shows that the number of ratings increased from 2013 to 2015 but then slightly decreased in recent years. This may indicate changes in user behavior or changes in the database's user base over time.

Regarding the distribution of ratings per movie, statistics show that most movies have an average rating around 3 to 4, suggesting that the majority of movies are well-received by users. However, there are also movies with extremely high or low ratings, suggesting that some movies may be particularly polarizing or exceptionally popular.

```python
# Function to check and summarize DataFrame
def summarize_dataframe(df):
    print("DataFrame Info:")
    print(df.info())
    print("\nDataFrame Description:")
    print(df.describe())
    print("\nTop 5 Rows:")
    print(df.head())

# Summarize movies DataFrame
print("\nMovies DataFrame:")
summarize_dataframe(df_movies)

# Summarize ratings DataFrame
print("\nRatings DataFrame:")
summarize_dataframe(df_ratings)
```

# Model

The code for a movie recommendation system uses collaborative filtering to provide recommendations based on user ratings. By first importing necessary libraries such as pandas, numpy, and NearestNeighbors from sklearn, and loading movie information and user ratings from CSV files, the data is prepared for analysis. By filtering out less popular movies and creating a sparse matrix using the pivot_table function in Pandas, user ratings are converted into a CSR matrix suitable for calculating nearest neighbors. Then, a NearestNeighbors model is created and fit to the created matrix to compute the nearest neighbors for each movie.

# Results
